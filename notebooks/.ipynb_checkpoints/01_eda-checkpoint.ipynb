## 2. Data Cleaning

```python
# Create a copy of raw data
df_clean = df.copy()
```

```python
# Convert date columns to datetime
df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'])
df_clean['Ship Date'] = pd.to_datetime(df_clean['Ship Date'])
```

```python
# Convert Zip Code to String
df_clean['Postal Code'] = df_clean['Postal Code'].astype(str)
```

```python
# Standardize Text Cols
text_cols = ['Ship Mode','Segment','Country','City','State','Region','Category','Sub-Category']

for col in text_cols:
    df_clean[col] = df_clean[col].str.strip().str.title()
    
    
print(f"Shape: {df_clean.shape}")
```

```python
df_clean.head()
```

## 3. Feature Engineering

```python
# Extract date features from order date
df_clean['Order Year'] = df_clean['Order Date'].dt.year
df_clean['Order Month'] = df_clean['Order Date'].dt.month
df_clean['Order Month Name'] = df_clean['Order Date'].dt.strftime('%B')
df_clean['Order Quarter'] = df_clean['Order Date'].dt.quarter

# We can figure out how many days it took to ship an item
df_clean['Days to Ship'] = (df_clean['Ship Date'] - df_clean['Order Date']).dt.days

# We can also find profit margin per order
df_clean['Profit Margin Percentage'] = ((df_clean['Profit'] / df_clean['Sales']) * 100).round(2)
```

```python
print(f"New Shape: {df_clean.shape}")
print(f"\nNew Columns Added:")
print(['Order Year', 'Order Month','Order Month Name', 'Order Quarter', 'Days to Ship','Profit Margin Percentage'])
```

## 4. Verify Data

```python
# Verify Data is Correct
print("Order Date Range")
print(f"    Earliest  : {df_clean['Order Date'].min().date()}")
print(f"    Latest   : {df_clean['Order Date'].max().date()}")
```

```python
# Check for Negative Shipping Days
neg_ship = (df_clean['Days to Ship'] < 0).sum()
print(f"\nNegative Shipping Days: {neg_ship}")
```

```python
# Lets look at Days To Ship summary
print(f"\nDays to Ship Summary:")
print(df_clean['Days to Ship'].describe().round(2))
```
